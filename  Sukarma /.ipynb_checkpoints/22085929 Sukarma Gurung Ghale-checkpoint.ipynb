Write a python program to find out top 15 jobs. Make a bar graph of sales as well.

```python
# Counting the occurrences of 
#each job title and take the top 15 most frequent ones
top_jobs=file['job_title'].value_counts().head(15)
top_jobs


```

```python
# Create bar graph
plt.bar(top_jobs.index, top_jobs.values, color='blue')
plt.xlabel('Job Title')
plt.ylabel('Total Salary in USD')
plt.title('Top 15 Jobs by Sales')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

Which job has the highest salaries? Illustrate with bar graph.


```python
# Group data by 'job_title' and calculate the mean salary for each group,
# then sort the result in descending order
mean_salary_by_job = file.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

# Find job with the highest mean salary using idxmax() function, which 
#returns the index of the maximum value
highest_salary_job = mean_salary_by_job.idxmax()
highest_salary_job
```

```python
# Set the figure size to 25x6 inches
plt.figure(figsize=(25, 6))
# Create a bar plot using the mean salary data, with pink color
mean_salary_by_job.plot(kind='bar', color='pink')
# Set title for the plot
plt.title('Mean Salary by Job Title')
# Set label for x-axis
plt.xlabel('Job Title')
# Set label for y-axis
plt.ylabel('Mean Salary (in USD)')
# Rotate x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)
# Display the legend to show the mapping between bar colors and job titles
plt.legend()
# Display the plot
plt.show()
```

Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
# Group data by 'experience_level' and calculate the mean salary for each group,
# then sort the result in descending order
salary_by_experience = file.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
plt.figure(figsize=(5,6))
# Create a bar plot using the mean salary data, with sky blue color
salary_by_experience.plot(kind='bar',color='skyblue')
# Set title for the plot
plt.title(' Mean Salary by Experience Level')
# Set label for x-axis
plt.xlabel('Experience Level')
# Set label for y-axis
plt.ylabel('Mean Salary (in USD)')
# Rotate x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)
# Adjust layout to prevent overlap of labels
plt.tight_layout()
# Display the plot
plt.show()
```

Write a Python program to show histogram and box plot of any chosen
different variables. Use proper labels in the graph.

```python
#Function to generate a histogram and box plot for a given variable.

def histogram_and_boxplot(data, variable_name):
   
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)  # Create a subplot for histogram
    plt.hist(data, bins=20, color='pink', edgecolor='black')
    plt.title('Histogram of ' + variable_name)
    plt.xlabel(variable_name)
    plt.ylabel('Frequency')
# Create a subplot for box plot
    plt.subplot(1, 2, 2) 
    plt.boxplot(data, vert=False)
    plt.title('Box Plot of ' + variable_name)
    plt.xlabel(variable_name)
    plt.yticks([])  # Hide y-axis ticks
     # Adjust layout to prevent overlap of subplots
    plt.tight_layout()
# Display the plots
    plt.show()

# Example usage
variable_name = 'salary_in_usd'
# Call the function with the data and variable name
histogram_and_boxplot(file[variable_name], variable_name)
```
